```python
import numpy as np
import matplotlib.pyplot as plt

# Define the time parameters
t0 = 0
t_end = 10
delta_t = 0.1
time_steps = int((t_end - t0) / delta_t) + 1

# Initialize arrays to store the state variables
x = np.zeros(time_steps)
y = np.zeros(time_steps)
q = np.zeros(time_steps)

# Set initial conditions
x[0] = 0
y[0] = 0
q[0] = 1


# Define control inputs
def wt(t):
    if 4 <= t < 6:
        return 12
    elif 6 <= t <= 8:
        return 12
    else:
        return 1


def wr(t):
    if 0.5 <= t <= 1.5:
        return 12
    elif 2 <= t <= 4:
        return 12
    else:
        return 1


# Simulate the unicycle robot model using Euler's method
for i in range(1, time_steps):
    t = i * delta_t
    x_dot = q[i - 1] * np.cos(q[i - 1]) * wr(t)
    y_dot = q[i - 1] * np.sin(q[i - 1]) * wr(t)
    q_dot = wt(t)

    x[i] = x[i - 1] + delta_t * x_dot
    y[i] = y[i - 1] + delta_t * y_dot
    q[i] = q[i - 1] + delta_t * q_dot

# Create time array
time = np.linspace(t0, t_end, time_steps)

# Plot x vs y
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='x vs y')
plt.xlabel('x')
plt.ylabel('y')
plt.title('x vs y')
plt.grid()
plt.legend()
plt.show()

# Plot x vs t
plt.figure(figsize=(10, 5))
plt.plot(time, x, label='x vs t')
plt.xlabel('t')
plt.ylabel('x')
plt.title('x vs t')
plt.grid()
plt.legend()
plt.show()

# Plot y vs t
plt.figure(figsize=(10, 5))
plt.plot(time, y, label='y vs t')
plt.xlabel('t')
plt.ylabel('y')
plt.title('y vs t')
plt.grid()
plt.legend()
plt.show()

# Plot q vs t
plt.figure(figsize=(10, 5))
plt.plot(time, q, label='q vs t')
plt.xlabel('t')
plt.ylabel('q')
plt.title('q vs t')
plt.grid()
plt.legend()
plt.show()
```
