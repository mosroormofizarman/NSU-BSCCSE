```python
import numpy as np
import matplotlib.pyplot as plt

# Define the time step and time span
dt = 0.1
t = np.arange(0, 10 + dt, dt)

# Initialize arrays to store the state variables
x = np.zeros(len(t))
y = np.zeros(len(t))
q = np.zeros(len(t))

# Initial conditions
x[0] = 0
y[0] = 0
q[0] = 1


# Define control input functions
def v(t):
    return 1


def omega(t):
    if 0.5 <= t < 1.5:
        return 3
    elif 2 <= t < 3:
        return -3
    elif 4 <= t < 5:
        return -3
    elif 6 <= t < 7:
        return 3
    elif 8 <= t < 9:
        return -3
    else:
        return 0


# Simulate the unicycle robot model using Euler's method
for i in range(1, len(t)):
    x_dot = v(t[i - 1]) * np.cos(q[i - 1])
    y_dot = v(t[i - 1]) * np.sin(q[i - 1])
    q_dot = omega(t[i - 1])

    x[i] = x[i - 1] + x_dot * dt
    y[i] = y[i - 1] + y_dot * dt
    q[i] = q[i - 1] + q_dot * dt

# Plotting
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title('x vs y')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2, 2, 2)
plt.plot(t, x)
plt.title('x vs t')
plt.xlabel('t')
plt.ylabel('x')

plt.subplot(2, 2, 3)
plt.plot(t, y)
plt.title('y vs t')
plt.xlabel('t')
plt.ylabel('y')

plt.subplot(2, 2, 4)
plt.plot(t, q)
plt.title('q vs t')
plt.xlabel('t')
plt.ylabel('q')

plt.tight_layout()
plt.show()
```
