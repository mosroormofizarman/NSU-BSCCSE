```python
#2c
import numpy as np
import matplotlib.pyplot as plt

# Define initial conditions
x0, y0 = 0, 0
V0 = 0.5
theta0 = -np.pi/2  # Initial orientation in radians

# Define time parameters
dt_values = [0.1, 0.01]  # Time step sizes
tf = 15  # Final time

# Define control functions (example constants for now)
def control_a(t):
    return 0.1  # Example constant acceleration

def control_w(t):
    return 0.01  # Example constant angular velocity

# Function to integrate using Euler's method
def integrate(dt):
    # Initialize arrays to store trajectory
    t_values = np.arange(0, tf, dt)
    x_values = np.zeros_like(t_values)
    y_values = np.zeros_like(t_values)
    V_values = np.zeros_like(t_values)
    theta_values = np.zeros_like(t_values)

    # Set initial conditions
    x_values[0] = x0
    y_values[0] = y0
    V_values[0] = V0
    theta_values[0] = theta0

    # Integrate using Euler's method
    for i in range(1, len(t_values)):
        # Update controls
        a_i = control_a(t_values[i-1])
        w_i = control_w(t_values[i-1])
        
        # Update state using Euler's method
        x_values[i] = x_values[i-1] + dt * V_values[i-1] * np.cos(theta_values[i-1])
        y_values[i] = y_values[i-1] + dt * V_values[i-1] * np.sin(theta_values[i-1])
        V_values[i] = V_values[i-1] + dt * a_i
        theta_values[i] = theta_values[i-1] + dt * w_i

    return t_values, x_values, y_values

# Plot the results
plt.figure(figsize=(12, 6))

for dt in dt_values:
    t, x, y = integrate(dt)
    plt.plot(x, y, label=f'Δt = {dt}')

plt.title('Robot Trajectory (Euler\'s Method)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axis('equal')
plt.legend()
plt.show()
```
