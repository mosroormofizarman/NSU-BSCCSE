```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define control inputs
def control_input(t):
  if 0 <= t <= 10:
    v = 1
  else:
    v = 0
  if 0.5 <= t <= 1.5 or 6 <= t <= 7:
    w = 3
  elif 2 <= t <= 3 or 4 <= t <= 5 or 8 <= t <= 9:
    w = -3
  else:
    w = 0
  return v, w
```

```python
# Define differential equations
def f(x, y, theta, v, w):
  x_dot = v * np.cos(theta)
  y_dot = v * np.sin(theta)
  theta_dot = w
  return x_dot, y_dot, theta_dot
```

```python
# Simulation parameters
dt = 0.1
T = 10

# Initial conditions
x = 0
y = 0
theta = 1
```

```python
# Initialize data storage
t_data = np.arange(0, T + dt, dt)
x_data = np.zeros(len(t_data))
y_data = np.zeros(len(t_data))
theta_data = np.zeros(len(t_data))
```

```python
# Euler's method loop
for i in range(len(t_data)):
  v, w = control_input(t_data[i])
  x_dot, y_dot, theta_dot = f(x, y, theta, v, w)
  x += x_dot * dt
  y += y_dot * dt
  theta += theta_dot * dt
  x_data[i] = x
  y_data[i] = y
  theta_data[i] = theta
```

```python
# Plot the results
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
plt.plot(x_data, y_data)
plt.xlabel('X (m)')
plt.ylabel('Y (m)')
plt.title('Robot trajectory')

plt.subplot(2, 2, 2)
plt.plot(t_data, x_data)
plt.xlabel('Time (s)')
plt.ylabel('X (m)')
plt.title('X vs time')

plt.subplot(2, 2, 3)
plt.plot(t_data, y_data)
plt.xlabel('Time (s)')
plt.ylabel('Y (m)')
plt.title('Y vs time')

plt.subplot(2, 2, 4)
plt.plot(t_data, theta_data)
plt.xlabel('Time (s)')
plt.ylabel('Theta (rad)')
plt.title('Theta vs time')

plt.tight_layout()
plt.show()
```
