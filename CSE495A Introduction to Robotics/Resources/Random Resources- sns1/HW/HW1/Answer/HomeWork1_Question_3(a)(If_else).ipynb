```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define constants
r = 0.1  # Wheel radius (m)
L = 1     # Distance between wheels (m)
dt = 0.1  # Time step (s)
t_max = 10 # Maximum simulation time (s)
```

```python
# Define control inputs as functions of time
def wl(t):
    if 4 <= t <= 6:
        return 12
    elif 6 <= t <= 8:
        return 12
    else:
        return 1

def wr(t):
    if 0.5 <= t <= 1.5:
        return 12
    elif 2 <= t <= 4:
        return 12
    else:
        return 1
```

```python
# Initialize state variables
x = 0
y = 0
theta = 1
```

```python
# Initialize empty lists for storing data
t_data = []
x_data = []
y_data = []
theta_data = []
```

```python
# Simulate using Euler's method
for t in np.arange(0, t_max + dt, dt):
    # Update state variables using control inputs
    v = (wl(t) + wr(t)) * r / 2
    w = (wr(t) - wl(t)) * r / L

    x_dot = v * np.cos(theta)
    y_dot = v * np.sin(theta)
    theta_dot = w

    x += x_dot * dt
    y += y_dot * dt
    theta += theta_dot * dt

    # Store data for plotting
    t_data.append(t)
    x_data.append(x)
    y_data.append(y)
    theta_data.append(theta)
```

```python
# Plot results
plt.figure(figsize=(10, 6))

plt.subplot(221)
plt.plot(x_data, y_data)
plt.xlabel('X (m)')
plt.ylabel('Y (m)')
plt.title('Robot Trajectory')

plt.subplot(222)
plt.plot(t_data, x_data)
plt.xlabel('Time (s)')
plt.ylabel('X (m)')
plt.title('X vs Time')

plt.subplot(223)
plt.plot(t_data, y_data)
plt.xlabel('Time (s)')
plt.ylabel('Y (m)')
plt.title('Y vs Time')

plt.subplot(224)
plt.plot(t_data, theta_data)
plt.xlabel('Time (s)')
plt.ylabel('Theta (rad)')
plt.title('Theta vs Time')

plt.tight_layout()
plt.show()
```
