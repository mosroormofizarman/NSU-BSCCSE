```python
import numpy as np
import matplotlib.pyplot as plt
```

# Initialization Variable

```python
Tinitial = 0
Tfinal = 10
delT = 0.1
```

```python
arraySize = int ((Tfinal - Tinitial)/delT)
arraySize
```

```python
x = np.zeros(int(arraySize))
y = np.zeros(int(arraySize))
theta = np.ones(int(arraySize))
x, y , theta
```

```python
r = 0.1
l = 1
wl = np.ones(100)
wr = np.ones(100)
wl , wr
```

# Condition


```python
for i in range(100):

  wr[i] = 1
  if i>=5 and i<=15:
    wr[i] = 12
  elif i>=20 and i<=40:
    wr[i] = 12
  elif i>=40 and i<=60:
    wl[i] = 12
  elif i>=60 and i<=80:
    wl[i] = 12


wl, wr
```

# Main Program

```python
for i in range(99):
  x[i+1] =  x[i] + ((r/2) *(wl[i]+wr[i])*(np.cos(theta[i]))) *delT
  y[i+1] =  y[i] + ((r/2) *(wl[i]+wr[i])*(np.sin(theta[i]))) * delT
  theta[i+1] =  theta[i] + ((r/l) *(wr[i]-wl[i]))* delT


x, y, theta
```

```python
t= np.arange(100)

plt.plot(t, x)
plt.title('X vs T')
plt.xlabel('X')
plt.ylabel('T')
```

```python
plt.plot(t, y)
plt.title('Y vs T')
plt.xlabel('Y')
plt.ylabel('T')
```

```python
plt.plot(t, theta)
plt.title('θ vs T')
plt.xlabel('θ')
plt.ylabel('T')
```

```python
plt.plot(x,y)
plt.title('X vs Y')
plt.xlabel('X')
plt.ylabel('Y')
```
